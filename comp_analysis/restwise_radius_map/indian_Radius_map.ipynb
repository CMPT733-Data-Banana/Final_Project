```python
import pandas as pd
import folium
```

```python
#!pip3 install haversine
from haversine import haversine
```

```python
att = pd.read_csv('Downloads/attributes.csv')
```

```python
mb3 = pd.read_csv("rest5og.csv")
mbc3 = pd.read_csv("rest5.csv")

latitude_list4 = mbc3['latitude'].astype(float)

longitude_list4 = mbc3['longitude'].astype(float)
longitude_list4

latitude_list3 = mb3['latitude'].astype(float)

longitude_list3 = mb3['longitude'].astype(float)

latitude_list3

len(latitude_list3)

len(latitude_list4)

dist = []

for i in range(len(latitude_list4)):
    disty = haversine((latitude_list4[i], longitude_list4[i]),
                      (latitude_list3, longitude_list3),
                      unit='km')
    dist.append(disty)

dist

mbc3['dists'] = dist

mbc3_filtered = mbc3[mbc3['dists'] < 5]

mb3_att = att[att['business_id'].isin(mb3['business_id'])]

merged_source_mb3 = pd.merge(mb3_att, mb3, on='business_id')

mbc3_att_df = att[att['business_id'].isin(mbc3['business_id'])]

merged_mb3 = pd.merge(mbc3_att_df, mbc3_filtered, on='business_id')

merged_mb3 = merged_mb3.drop(columns=['postal_code', 'categories', 'stars_x'])

merged_mb3_filtered = merged_mb3[
    merged_mb3['attributes.RestaurantsPriceRange2'] == 2]

merged_mb3_filtered = merged_mb3_filtered.sort_values(by='dists')

merged_mb3_filtered_1km = merged_mb3_filtered[
    merged_mb3_filtered['dists'] <= 1]

merged_mb3_filtered_2km = merged_mb3_filtered[
    merged_mb3_filtered['dists'] <= 2]

merged_mb3_filtered_3km = merged_mb3_filtered[
    merged_mb3_filtered['dists'] <= 3]

map2 = folium.Map(location=[49.285006, -123.118050],
                  tiles='OpenStreetMap',
                  zoom_start=12)

m2fg = folium.FeatureGroup(
    name='SULA Restuarant and its competiton in 1km radius', overlay=True)
m22fg = folium.FeatureGroup(
    name='SULA Restuarant and its competiton in 2km radius', overlay=True)

m23fg = folium.FeatureGroup(
    name='SULA Restuarant and its competiton in 3km radius', overlay=True)

m25fg = folium.FeatureGroup(
    name='SULA Restuarant and its competiton in 5km radius', overlay=True)

m2fg.add_to(map2)
m22fg.add_to(map2)
m23fg.add_to(map2)

m25fg.add_to(map2)

#layer for 1
for index, row in merged_source_mb3.iterrows():
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=' Reviews:' + str(row['stars_y']) + '\t' + ' Name' + ':' +
        str(row['name']) + ' Wifi:' + str(row['attributes.WiFi']) +
        ' Price Range:' + str(row['attributes.RestaurantsPriceRange2']) + '$' +
        ' Delivery:' + str(row['attributes.RestaurantsDelivery']),
        icon=folium.Icon(color='darkgreen', icon='glass')).add_to(m2fg)
    folium.Circle([row['latitude'], row['longitude']],
                  radius=3000,
                  label='5KM Radius',
                  color='cadetblue').add_to(m2fg)

a = merged_mb3_filtered_1km['latitude'].to_list()
b = merged_mb3_filtered_1km['longitude'].to_list()
points = []

c = merged_source_mb3['latitude'].to_list()
d = merged_source_mb3['longitude'].to_list()
pointsc = []

for i in range(len(a)):
    points.append([a[i], b[i]])

for i in range(len(c)):
    pointsc.append([c[i], d[i]])

for index, row in merged_mb3_filtered_1km.iterrows():

    iframe = folium.IFrame(width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=' Reviews:' + str(row['stars_y']) + '\t' + ' Name' + ':' +
        str(row['name']) + ' Wifi:' + str(row['attributes.WiFi']) +
        ' Price Range:' + str(row['attributes.RestaurantsPriceRange2']) + '$' +
        ' Delivery' + ':' + str(row['attributes.RestaurantsDelivery']),
        icon=folium.Icon(color='darkred', icon='glass')).add_to(m2fg)

#layer for 2

for index, row in merged_source_mb3.iterrows():
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=' Reviews:' + str(row['stars_y']) + '\t' + ' Name' + ':' +
        str(row['name']) + ' Wifi:' + str(row['attributes.WiFi']) +
        ' Price Range:' + str(row['attributes.RestaurantsPriceRange2']) + '$' +
        ' Delivery:' + str(row['attributes.RestaurantsDelivery']),
        icon=folium.Icon(color='darkgreen', icon='glass')).add_to(m22fg)
    folium.Circle([row['latitude'], row['longitude']],
                  radius=3000,
                  label='5KM Radius',
                  color='cadetblue').add_to(m22fg)

a = merged_mb3_filtered_2km['latitude'].to_list()
b = merged_mb3_filtered_2km['longitude'].to_list()
points = []

c = merged_source_mb3['latitude'].to_list()
d = merged_source_mb3['longitude'].to_list()
pointsc = []

for i in range(len(a)):
    points.append([a[i], b[i]])

for i in range(len(c)):
    pointsc.append([c[i], d[i]])

for index, row in merged_mb3_filtered_2km.iterrows():

    iframe = folium.IFrame(width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=' Reviews:' + str(row['stars_y']) + '\t' + 'Name' + ':' +
        str(row['name']) + ' Wifi:' + str(row['attributes.WiFi']) +
        'Price Range:' + str(row['attributes.RestaurantsPriceRange2']) + '$' +
        'Delivery' + ':' + str(row['attributes.RestaurantsDelivery']),
        icon=folium.Icon(color='darkred', icon='glass')).add_to(m23fg)

#layer for 3

for index, row in merged_source_mb3.iterrows():
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=' Reviews:' + str(row['stars_y']) + '\t' + ' Name' + ':' +
        str(row['name']) + ' Wifi:' + str(row['attributes.WiFi']) +
        ' Price Range:' + str(row['attributes.RestaurantsPriceRange2']) + '$' +
        ' Delivery:' + str(row['attributes.RestaurantsDelivery']),
        icon=folium.Icon(color='darkgreen', icon='glass')).add_to(m23fg)
    folium.Circle([row['latitude'], row['longitude']],
                  radius=3000,
                  label='3KM Radius',
                  color='cadetblue').add_to(m23fg)

a = merged_mb3_filtered_3km['latitude'].to_list()
b = merged_mb3_filtered_3km['longitude'].to_list()
points = []

c = merged_source_mb3['latitude'].to_list()
d = merged_source_mb3['longitude'].to_list()
pointsc = []

for i in range(len(a)):
    points.append([a[i], b[i]])

for i in range(len(c)):
    pointsc.append([c[i], d[i]])

for index, row in merged_mb3_filtered_3km.iterrows():

    iframe = folium.IFrame(width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=' Reviews:' + str(row['stars_y']) + '\t' + ' Name' + ':' +
        str(row['name']) + ' Wifi:' + str(row['attributes.WiFi']) +
        ' Price Range:' + str(row['attributes.RestaurantsPriceRange2']) + '$' +
        ' Delivery' + ':' + str(row['attributes.RestaurantsDelivery']),
        icon=folium.Icon(color='darkred', icon='glass')).add_to(m23fg)

#layer for 5
for index, row in merged_source_mb3.iterrows():
    folium.Marker([row['latitude'], row['longitude']],
                  tooltip=' Reviews:' + str(row['stars_y']) + '\t' + ' Name' +
                  ':' + str(row['name']),
                  icon=folium.Icon(color='darkgreen',
                                   icon='glass')).add_to(m25fg)
    folium.Circle([row['latitude'], row['longitude']],
                  radius=5000,
                  label='5KM Radius',
                  color='cadetblue').add_to(m25fg)

a = merged_mb3_filtered['latitude'].to_list()
b = merged_mb3_filtered['longitude'].to_list()
points = []

c = mb3['latitude'].to_list()
d = mb3['longitude'].to_list()
pointsc = []

for i in range(len(a)):
    points.append([a[i], b[i]])

for i in range(len(c)):
    pointsc.append([c[i], d[i]])

for index, row in merged_mb3_filtered.iterrows():

    iframe = folium.IFrame(width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker([row['latitude'], row['longitude']],
                  tooltip='Reviews: ' + str(row['stars_y']) + '\t' +
                  'Name ' + ':' + str(row['name']),
                  icon=folium.Icon(color='darkred',
                                   icon='glass')).add_to(m25fg)

folium.LayerControl(collapsed=False).add_to(map2)

print(
    'This map shows the resturants and their competitors where each marker indicates geocoordinates.'
)
print(
    'On 5km Radius, you can see name and ratings of resturants, further drilling down the radius helps you see common attributes between resturants and their competition.'
)
map2
```
