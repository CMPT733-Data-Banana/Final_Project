```python
import pandas as pd
```

```python
from wordcloud import STOPWORDS
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
stop_list= list(stopwords.words('english'))
```

```python
df_closed = pd.read_csv("reviews_closed.csv")
```

```python
df_open = pd.read_csv("reviews_open.csv")
```

```python
df_open
```

```python
print(stop_list)
```

```python
from wordcloud import WordCloud as wc

wordcloud_object = wc(background_color="white",stopwords = stop_list, 
                      min_font_size = 5, contour_width=3,width=600, height=400,
                      colormap='rainbow').generate(','.join(list(df_closed.text.values)))
wordcloud_object.to_file("Closed_reviews.png")
wordcloud_object.to_image()
```

```python
df_open['ratings'] = ''
df_open.loc[df_open.stars >=4, 'ratings'] = 'positive'
df_open.loc[df_open.stars ==3, 'ratings'] = 'neutral'
df_open.loc[df_open.stars <3, 'ratings'] = 'negative'

```

```python
from wordcloud import WordCloud as wc

wordcloud_object = wc(background_color="white",stopwords = stop_list, 
                      min_font_size = 5, contour_width=3,width=600, height=400,
                      colormap='rainbow').generate(','.join(list(df_open[df_open.ratings=="negative"].text.values)))

wordcloud_object.to_file("Negative_reviews.png")
wordcloud_object.to_image()
```
