```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df_closed = pd.read_csv("reviews_closed.csv")
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
import time
import datetime
```

```python
df_closed['date'] = pd.to_datetime(df_closed['date'])
df_closed = df_closed.set_index('date')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_score = df_closed['text'].map(SentimentIntensityAnalyzer().polarity_scores)
sent = sent_score.map(lambda x: x['compound'])
close_monthly_sentiment = sent.resample('M').mean()
```

```python
plt.plot(df_closed['text'].resample('M').count())
plt.xlabel('Year')
plt.ylabel('Sentiment of the Restaurant ')
plt.title('Average Monthly Sentiment')
plt.show()
```
